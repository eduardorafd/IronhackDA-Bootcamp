## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
orders2 = orders.groupby('CustomerID').sum()['amount_spent'].reset_index()
orders2
```

```python
Q95 = orders2['amount_spent'].quantile(0.95)
Q75 = orders2['amount_spent'].quantile(0.75)
```

```python
def customers(x):
    if x['amount_spent'] >= Q95:
        return ('VIP')
    elif x['amount_spent'] >= Q75 and x['amount_spent'] < Q95:
        return ('PREFERRED')
    return ('REGULAR')
```

```python
 orders2.apply(customers, axis=1)
```

```python
orders2['Group'] = orders2.apply(lambda x: customers(x), axis=1)
orders2
```

```python
customers_group = orders2[['CustomerID','Group']]
```

```python
customers_group
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
orders_customer = pd.merge(orders, customers_group, on='CustomerID', how='left')
```

```python
orders_customer
```

```python
columns = ['CustomerID','Country','Group']
```

```python
orders_customer = orders_customer.loc[:, columns]
orders_customer.head()
```

```python
orders_customer.pivot_table(
    values='CustomerID',
    index='Country',
    columns='Group',
    aggfunc='count'

).sort_values(by = 'VIP', ascending = False).head()
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
country_vip_pref = orders_customer[
    orders_customer['Group'].isin(['VIP','Preferred'])]
country_vip_pref
```

```python
orders_customers = country_vip_pref[[
    'Country', 'CustomerID']].groupby('Country').count()
orders_customers 
```

```python
orders_customers.sort_values(by = 'CustomerID', ascending = False).head()
```
