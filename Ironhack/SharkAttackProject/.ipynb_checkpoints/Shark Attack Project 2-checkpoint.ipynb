# SHARKS_CLEAN

```python
#Filters with loc
#my_project = sharks.loc[:, ['Case Number', 'Year', 'Type', 'Country', 'Area', 'Activity']]

#Filters
sharks = sharks.sort_values(by=['Country'])
my_project = sharks[['Year', 'Country', 'Location', 'Activity']]
my_project = my_project.dropna(subset=['Country'])
my_project = my_project.dropna(subset=['Year'])
#my_project = my_project.dropna(subset=['Location'])
my_project['Year'] = my_project['Year'].astype(int)
my_project = my_project.loc[my_project['Year'] > 2000 , :]
#String_Filters
my_project['Country'] = my_project['Country'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8') #Normaliz
my_project['Country'] = my_project['Country'].str.upper() #Upper
my_project['Country'] = my_project['Country'].str.replace(r"\(.*\)","") #Erase within parethesis
my_project['Country'] = my_project['Country'].replace('\?','', regex=True) #Replace Symbols
my_project['Country'] = my_project['Country'].replace ('/^[^ ][\w\W ]*[^ ]/', '', regex=True) #Eliminate startwith/endwith whitespaces
my_project = my_project.sort_values(by=['Country']) #Sort     
#RENAME_COUNTRIES
my_project['Country'] = my_project['Country'].replace({'ANDAMAN ISLANDS': 'ANDAMAN / NICOBAR ISLANDAS', 'CEYLON': 'SRI LANKA', 'CEYLON (SRI LANKA)': 'SRI LANKA'})
my_project['Country'] = my_project['Country'].replace({'EGYPT': 'EGYPT / ISRAEL', 'ITALY': 'ITALY / CROATIA'})
my_project['Country'] = my_project['Country'].replace({'RED SEA': 'RED SEA / INDIAN OCEAN'})
my_project['Country'] = my_project['Country'].replace({'REUNION': 'REUNION ISLAND','SOLOMON ISLANDS': 'SOLOMON ISLANDS / VANUATU', 'VANUATU': 'SOLOMON ISLANDS / VANUATU', 'ST HELENA, BRITISH OVERSEAS TERRITORY': 'ST HELENA'})
my_project['Country'] = my_project['Country'].replace({'ST. MAARTIN': 'ST. MARTIN', 'ST. MAARTEN': 'ST. MARTIN'})
my_project['Country'] = my_project['Country'].replace({'IRAN': 'IRAN / IRAQ', 'IRAQ': 'IRAN / IRAQ', 'MICRONESIA, FEDERATED STATES OF': 'MICRONESIA', 'FEDERATED STATES OF MICRONESIA' :  'MICRONESIA', 'CAMEROON' : 'EQUATORIAL GUINEA / CAMEROON', 'EQUATORIAL GUINEA' : 'EQUATORIAL GUINEA / CAMEROON'})
my_project['Country'] = my_project['Country'].replace({'MALDIVE ISLANDS': 'FALKLAND ISLANDS', 'MALDIVES ISLANDS' : 'FALKLAND ISLANDS', 'MALDIVES' : 'FALKLAND ISLANDS', 'WESTERN SAMOA' : 'SAMOA'})
my_project['Country'] = my_project['Country'].replace({'UNITED STATES OF AMERICA' : 'US', 'USA' : 'US', ' TONGA': 'TONGA', 'UNITED ARAB EMIRATES ': 'UNITED ARAB EMIRATES'})
                                                       
my_project
```

```python
#How many rows?
my_project['Country']
```

# Why?

### Why we have more attacks on Atlantic and USA?
1) Subnotification;

2) Tourism, sports and jobs;

3) Tourists tend to search warmer waters, which is Atlantic Coast for Americas and Pacific Coast for Oceania;

4) Seas Sizes: Pacific is much bigger and much less land, so less coast with uninhabited islands;

4) More populated regions like US has more Activity and then prone to incidents;

5) Shark species prone to attack closer to coasts;

6) Environmental imbalance where there is more human Activity can lead to shark attacks closer to coasts;
