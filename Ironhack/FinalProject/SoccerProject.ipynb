```python
url = 'https://sofifa.com/players?type=all&pn%5B0%5D=0&hl=en-US&showCol%5B%5D=pf&showCol%5B%5D=oa&showCol%5B%5D=pt&showCol%5B%5D=bp&showCol%5B%5D=vl&showCol%5B%5D=cr&showCol%5B%5D=fi&showCol%5B%5D=he&showCol%5B%5D=fr&showCol%5B%5D=lo&showCol%5B%5D=bl&showCol%5B%5D=st&showCol%5B%5D=ar&showCol%5B%5D=in&showCol%5B%5D=po&showCol%5B%5D=vi&showCol%5B%5D=pe&showCol%5B%5D=ma&showCol%5B%5D=sa&showCol%5B%5D=sl&showCol%5B%5D=gd&showCol%5B%5D=gh&showCol%5B%5D=gp&showCol%5B%5D=gr&showCol%5B%5D=dr'
```

```python
gk_names = []
gk_teams = []
gk_values = []
gk_overall = []
gk_potential = []
gk_foot = []
gk_bp = []
gk_positioning = []
gk_reflexes = []
gk_hand = []
gk_dive = []
gk_stand= []
gk_aggression = []
gk_marking = []
gk_slide = []
gk_intercept = []
gk_longpass = []
gk_stamina = []
gk_position = []
gk_penalty = []
gk_finish = []
gk_ballcontrol = []
gk_vision = []
gk_heading = []
gk_crossing = []
gk_dribbling = []
gk_freekick = []

for eduardo in tqdm(range(15)):
    response = requests.get(url + str(f'&offset={60*eduardo}'), 'lxml')
    html = response.content
    soup = BeautifulSoup(html)
    for a in soup.find_all('a', attrs = {'class': 'tooltip'}):
        gk_names.append(a.text)
        
    for b in soup.find_all('div', attrs = {'class': 'bp3-text-overflow-ellipsis'}):
            for c in b.find_all('a'):
                gk_teams.append(c.text)
                
    for d in soup.find_all('td', attrs = {'class': 'col col-vl'}):
        gk_values.append(d.text)  
            
    for e in soup.find_all('td', attrs = {'class': 'col col-oa col-sort'}):
        gk_overall.append(e.text)
            
    for f in soup.find_all('td', attrs = {'class': 'col col-pt'}):
        gk_potential.append(f.text)
            
    for g in soup.find_all('td', attrs = {'class': 'col col-pf'}):
        gk_foot.append(g.text) 
    
    for falt in soup.find_all('td', attrs = {'class': 'col col-bp'}):
        gk_bp.append(falt.text)
        
     # GOLEIROS   
        
    for h in soup.find_all('td', attrs = {'class': 'col col-gr'}):
        gk_reflexes.append(h.text) 
        
    for i in soup.find_all('td', attrs = {'class': 'col col-gp'}):
        gk_positioning.append(i.text) 
        
    for k in soup.find_all('td', attrs = {'class': 'col col-gh'}):
        gk_hand.append(k.text)
        
    for l in soup.find_all('td', attrs = {'class': 'col col-gd'}):
        gk_dive.append(l.text)
        
    # ZAGUEIROS     
        
    for t in soup.find_all('td', attrs = {'class': 'col col-sa'}):
        gk_stand.append(t.text) 
        
    for u in soup.find_all('td', attrs = {'class': 'col col-ar'}):
        gk_aggression.append(u.text) 
        
    for v in soup.find_all('td', attrs = {'class': 'col col-ma'}):
        gk_marking.append(v.text)
        
    for w in soup.find_all('td', attrs = {'class': 'col col-sl'}):
        gk_slide.append(w.text)
        
    for y in soup.find_all('td', attrs = {'class': 'col col-in'}):
        gk_intercept.append(y.text) 
        
    # LATERAIS 
        
    for tt in soup.find_all('td', attrs = {'class': 'col col-lo'}):
        gk_longpass.append(tt.text) 
        
    for vv in soup.find_all('td', attrs = {'class': 'col col-st'}):
        gk_stamina.append(vv.text)
        
    for ww in soup.find_all('td', attrs = {'class': 'col col-po'}):
        gk_position.append(ww.text)
        
    # RESTO - MEIAS , ATACANTES , VOLANTES 
    for j in soup.find_all('td', attrs = {'class': 'col col-fi'}):
        gk_finish.append(j.text) 
    
    for uu in soup.find_all('td', attrs = {'class': 'col col-pe'}):
        gk_penalty.append(uu.text) 
    
    for aa in soup.find_all('td', attrs = {'class': 'col col-bl'}):
        gk_ballcontrol.append(aa.text) 
        
    for bb in soup.find_all('td', attrs = {'class': 'col col-vi'}):
        gk_vision.append(bb.text) 
        
    for cc in soup.find_all('td', attrs = {'class': 'col col-he'}):
        gk_heading.append(cc.text) 
        
    for dd in soup.find_all('td', attrs = {'class': 'col col-fr'}):
        gk_freekick.append(dd.text) 
    
    for ee in soup.find_all('td', attrs = {'class': 'col col-cr'}):
        gk_crossing.append(ee.text) 
        
    for ff in soup.find_all('td', attrs = {'class': 'col col-dr'}):
        gk_dribbling.append(ff.text) 
```

```python
gk_names = list(filter(None, gk_names))
```

```python
gk_pd = pd.DataFrame({'Name' : gk_names, 'Teams': gk_teams, 'Values': gk_values, 'Position': gk_bp, 'Overall' : gk_overall, 'Potential': gk_potential, 'Foot' : gk_foot, 'Position' : gk_bp,
                      
                      'GK Positioning': gk_positioning, 'GK Reflexes': gk_reflexes, 'GK Handling': gk_hand, 'GK Diving': gk_dive,
                      
                      'Positioning': gk_position, 'Long Pass': gk_longpass, 'Stamina': gk_stamina, 'Penalties': gk_penalty,
                      
                      'Finishing': gk_finish, 'Ball Control': gk_ballcontrol, 'Vision': gk_vision, 'Heading': gk_heading,
                      
                      'Crossing': gk_crossing, 'Dribbling': gk_dribbling, 'Free Kick': gk_freekick,
                      
                      'Standing Tackle': gk_stand, 'Aggression': gk_aggression, 'Marking': gk_marking, 'Sliding': gk_slide, 'Intercept': gk_intercept})
```

```python
client = pd.DataFrame([[90, 90]], columns = ['Overall', 'Potential'])
def create_similarity(dataframe, client, metric='euclidean'):
    """
    Given a dataframe, return its similarity matrix 
    """
    distance_vector = cdist(client, dataframe[client.columns], metric)
    similarities = 1 / (1 + distance_vector)
    similarity_index = pd.DataFrame(similarities, 
                                    columns=dataframe.index, index = ['Recommendation']).T.sort_values('Recommendation', ascending = False)
    dataframe['Recommendation'] = similarity_index
    return dataframe.reset_index().sort_values('Recommendation', ascending = False)
```
