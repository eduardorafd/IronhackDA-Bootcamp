# Challenge 3: Hidden Cold Joke

Using Python, call Github API to find out the cold joke contained in the 24 secret files in the following repo:

https://github.com/ironhack-datalabs/scavenger

- The pieces of the secret are inside the files with .scavengerhunt in their name.  
- The files are in different directories of the repository scavenger.  
- The secret files are named from .0001.scavengerhunt to .0024.scavengerhunt. 
- They are scattered randomly throughout this repo. 
- You need to search for these files by calling the Github API, not searching the local files on your computer.

## Find the name of the directories

First, we use a search query to find the name of the directories where the secret files are stored.

```python
url2 = 'https://api.github.com/repos/ironhack-datalabs/scavenger/contents'
url3 = 'https://api.github.com/search/code?q=.scavengerhunt in:path+repo:ironhack-datalabs/scavenger'
x = requests.get(url3, auth=(user,token))
y = x.json()
y
```

## Find the path were the files are stored

```python
import re
data5 = []
for i in range(len(y['items'])):
    j = y['items'][i]['path']
    data5.append(j)
data5
def num_sort(test_string):
    return list(map(int, re.findall(r'\d+', test_string)))[1]
data5.sort(key=num_sort)
data5
```

## Get the texts inside the file

```python
content = 'https://api.github.com/repos/ironhack-datalabs/scavenger/contents/{+path}'
data6 = []
for i in data5:
    y2 = content.replace('{+path}', i)
    response = requests.get(y2, auth=(user,token))
    y3 = response.json()
    data6.append(y3)
data6
```

```python
data7 = []
for i in data6:
    data7.append(i['content'])
print(data7)
```

```python
import base64
final = []
for i in data7:
    base64_message = (i)
    base64_bytes = base64_message.encode('ascii')
    message_bytes = base64.b64decode(base64_bytes)
    message = message_bytes.decode('ascii')
    x = message.strip('\n')
    final.append(x)
print(final)
```

## Print the texts in the correct order

```python
' '.join(final)
```
