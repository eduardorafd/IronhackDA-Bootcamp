#### Question 1: Find the questions and answers of last month.

```python
# the time used in this API is in Unix epoch time!
# https://en.wikipedia.org/wiki/Unix_time
```

```python
SITE1 = StackAPI('stackoverflow')
from datetime import datetime
questions = SITE1.fetch('questions', fromdate=datetime(2021,3,14), todate=datetime(2021,4,14))
for i in questions['items']:
    print(i['title'])
```

```python
pd.DataFrame(questions['items'])['title']
```

```python
SITE2 = StackAPI('stackoverflow')
from datetime import datetime
answers = SITE2.fetch('answers', fromdate=datetime(2021,3,14), todate=datetime(2021,4,14))
answers['items']
```

```python
pd.DataFrame(answers['items'])
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
SITE2 = StackAPI('stackoverflow')
from datetime import datetime
questions = SITE2.fetch('questions', fromdate=datetime(2021,4,13), todate=datetime(2021,4,14), min = 5, sort = 'votes', tagged = 'python')
data = []
x = []
for i in questions['items']:
    data.append(i['title'])
    x.append(i['score'])
y = list(zip(data, x))
y
```

#### Question 3: Find the answers with id 6784 and 6473.

```python
SITE3 = StackAPI('stackoverflow')
from datetime import datetime
answers = SITE3.fetch('answers', ids = [6784, 6473])
answers
```
